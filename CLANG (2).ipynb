```python
from clang.cindex import Index, CursorKind
import sys

def parse_file(filename):
    index = Index.create()
    tu = index.parse(filename)

    print(f"✅ Clang parsed {filename} successfully!")

    # Filter function to skip system headers
    def is_user_code(cursor):
        return (cursor.location.file and
                cursor.location.file.name == filename)

    # Function to print relevant nodes
    def print_node(cursor, depth=0):
        if not is_user_code(cursor):
            return

        # Print OpenMP pragmas and function definitions
        if (cursor.kind == CursorKind.OPENMP_DIRECTIVE or
            cursor.kind == CursorKind.FUNCTION_DECL or
            cursor.kind == CursorKind.CALL_EXPR or
            cursor.kind == CursorKind.VAR_DECL):

            print("  " * depth + f"{cursor.kind}: {cursor.spelling}")

        # Recurse for children
        for child in cursor.get_children():
            print_node(child, depth + 1)

    print_node(tu.cursor)

# Test with your file
parse_file("P3.c")
```

```python
from clang.cindex import CursorKind

def extract_features(cursor):
    features = {
        'openmp_pragmas': 0,
        'parallel_regions': 0,
        'shared_vars': 0,
        'critical_sections': 0,
        'reduction_ops': 0,
        'race_conditions': 0,
        'deadlock_risks': 0,
        'for_loops': 0,
        'atomic_ops': 0
    }

    def analyze_node(node):
        # Get OpenMP directives and clauses
        if node.kind == CursorKind.OMP_PARALLEL_DIRECTIVE:
            features['parallel_regions'] += 1
            features['openmp_pragmas'] += 1

        elif node.kind == CursorKind.OMP_FOR_DIRECTIVE:
            features['openmp_pragmas'] += 1
            features['for_loops'] += 1

        elif node.kind == CursorKind.OMP_CRITICAL_DIRECTIVE:
            features['critical_sections'] += 1
            features['openmp_pragmas'] += 1

        elif node.kind == CursorKind.OMP_ATOMIC_DIRECTIVE:
            features['atomic_ops'] += 1
            features['openmp_pragmas'] += 1

        elif node.kind == CursorKind.OMP_REDUCTION_CLAUSE:
            features['reduction_ops'] += 1

        # Detect shared variables (potential race conditions)
        if node.kind == CursorKind.VAR_DECL:
            # Check if variable is shared in parallel region
            parent = node.semantic_parent
            while parent:
                if parent.kind == CursorKind.OMP_PARALLEL_DIRECTIVE:
                    features['shared_vars'] += 1
                    # If no atomic/critical protection, flag as race condition
                    if features['atomic_ops'] == 0 and features['critical_sections'] == 0:
                        features['race_conditions'] += 1
                    break
                parent = parent.semantic_parent

        # Detect deadlock risks
        if (node.kind == CursorKind.CALL_EXPR and
            node.spelling in ['omp_set_lock', 'omp_set_nest_lock']):
            features['deadlock_risks'] += 1

        # Recurse through children
        for child in node.get_children():
            analyze_node(child)

    analyze_node(cursor)
    return features
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def extract_tf_tfidf(file_path):
    with open(file_path, "r") as f:
        code_text = f.read()  # Read source code

    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([code_text])

    tf_features = tfidf_matrix.mean(axis=1).flatten()[0]  # Mean TF score
    tfidf_features = tfidf_matrix.max(axis=1).flatten()[0]  # Max TF-IDF score
    return tf_features, tfidf_features

tf_features, tfidf_features = extract_tf_tfidf("p2.c")
print("TF Features:", tf_features)
print("TF-IDF Features:", tfidf_features)
```

```python
import clang.cindex

# Initialize Clang index
index = clang.cindex.Index.create()

# Parse the test file
tu = index.parse("P3.c")
print("✅ Clang parsed the test file successfully!")

# Function to check if a node is from a system header
def is_system_header(node):
    return node.location.file and "P3.c" not in str(node.location.file)

# Extract OpenMP pragmas manually
def extract_pragmas(file_path):
    pragmas = []
    with open(file_path, "r") as f:
        for line_number, line in enumerate(f, start=1):
            if "#pragma omp" in line:
                pragmas.append((line_number, line.strip()))
    return pragmas

# Detect potential shared variables in parallel regions
def detect_shared_variables(tu):
    shared_variables = set()
    in_parallel_region = False

    for node in tu.cursor.walk_preorder():
        if is_system_header(node):
            continue

        # Check if we are inside an OpenMP parallel region
        if node.kind == clang.cindex.CursorKind.COMPOUND_STMT and "#pragma omp parallel" in str(node.extent):
            in_parallel_region = True

        # Identify variables modified within parallel regions
        if in_parallel_region and node.kind == clang.cindex.CursorKind.VAR_DECL:
            shared_variables.add(node.spelling)

        # Exit parallel region
        if in_parallel_region and node.kind == clang.cindex.CursorKind.COMPOUND_STMT:
            in_parallel_region = False

    return shared_variables

# Extract and print OpenMP pragmas
pragmas = extract_pragmas("P3.c")
if pragmas:
    print("\n🔍 Found OpenMP pragmas:")
    for line_number, pragma in pragmas:
        print(f"Line {line_number}: {pragma}")
else:
    print("\n🔍 No OpenMP pragmas found.")

# Traverse the AST and print relevant nodes
print("\n📋 Relevant nodes in the AST:")
for node in tu.cursor.walk_preorder():
    # Skip nodes from system headers
    if is_system_header(node):
        continue

    # Print relevant nodes (e.g., functions, variables, loops)
    if node.kind in [
        clang.cindex.CursorKind.FUNCTION_DECL,
        clang.cindex.CursorKind.VAR_DECL,
        clang.cindex.CursorKind.PARM_DECL,
        clang.cindex.CursorKind.CALL_EXPR,
        clang.cindex.CursorKind.COMPOUND_STMT,
        clang.cindex.CursorKind.FOR_STMT
    ]:
        print(f"{node.kind} - {node.spelling}")

# Detect potential race conditions by identifying shared variables
shared_variables = detect_shared_variables(tu)
if shared_variables:
    print("\n⚠️ Potential shared variables in parallel regions:")
    for var in shared_variables:
        print(f"- {var}")
else:
    print("\n⚠️ No shared variables detected in parallel regions.")
```

```python
import clang.cindex
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize Clang index
index = clang.cindex.Index.create()

# List of source files and their labels (1 = defective, 0 = non-defective)
files_and_labels = [
    ("P3.c", 1),  # Defective file (contains a race condition)
    ("p4.c", 1),  # Defective file (high overhead)
    ("p5.c", 0),  # Non-defective file
    # Add more files here...
]

# Tokenize and extract features for each file
all_token_texts = []
labels = []

for file, label in files_and_labels:
    # Parse the file
    tu = index.parse(file)

    # Get the main source file
    source_file = tu.get_file(file)
    if not source_file:
        raise ValueError(f"Could not find the source file '{file}'.")

    # Read the file contents to calculate the number of lines
    with open(file, "r") as f:
        file_contents = f.read()
    num_lines = file_contents.count("\n") + 1  # Add 1 to include the last line

    # Define the extent of the file (from the start to the end)
    start = clang.cindex.SourceLocation.from_position(tu, source_file, line=1, column=1)
    end = clang.cindex.SourceLocation.from_position(tu, source_file, line=num_lines, column=1)
    source_range = clang.cindex.SourceRange.from_locations(start, end)

    # Extract tokens within the defined source range
    tokens = tu.get_tokens(extent=source_range)
    token_list = []

    for token in tokens:
        if file in str(token.location.file):  # Skip system headers
            token_list.append(token.spelling)

    # Combine tokens into a single string
    token_text = " ".join(token_list)
    all_token_texts.append(token_text)
    labels.append(label)

# Convert token texts into TF-IDF features
tfidf_vectorizer = TfidfVectorizer()
tfidf_features = tfidf_vectorizer.fit_transform(all_token_texts)

print("\n📊 Combined TF-IDF Features Shape:", tfidf_features.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(tfidf_features, labels, test_size=0.2, random_state=42)

# Train Random Forest
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

# Train SVM
svm = SVC(kernel='linear', random_state=42)
svm.fit(X_train, y_train)

# Evaluate Random Forest
y_pred_rf = rf.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)

# Evaluate SVM
y_pred_svm = svm.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
precision_svm = precision_score(y_test, y_pred_svm)
recall_svm = recall_score(y_test, y_pred_svm)
f1_svm = f1_score(y_test, y_pred_svm)

print("\n🤖 Random Forest Metrics:")
print(f"Accuracy: {accuracy_rf * 100:.2f}%")
print(f"Precision: {precision_rf * 100:.2f}%")
print(f"Recall: {recall_rf * 100:.2f}%")
print(f"F1 Score: {f1_rf * 100:.2f}%")

print("\n🤖 SVM Metrics:")
print(f"Accuracy: {accuracy_svm * 100:.2f}%")
print(f"Precision: {precision_svm * 100:.2f}%")
print(f"Recall: {recall_svm * 100:.2f}%")
print(f"F1 Score: {f1_svm * 100:.2f}%")
```

```python
from sklearn.model_selection import cross_val_score

# Perform 10-fold cross-validation for Random Forest
cv_scores_rf = cross_val_score(rf, tfidf_features, labels, cv=10, scoring="accuracy")
print("\n🎯 Random Forest Cross-Validation Accuracy:", f"{cv_scores_rf.mean() * 100:.2f}% ± {cv_scores_rf.std() * 100:.2f}%")

# Perform 10-fold cross-validation for SVM
cv_scores_svm = cross_val_score(svm, tfidf_features, labels, cv=10, scoring="accuracy")
print("\n🎯 SVM Cross-Validation Accuracy:", f"{cv_scores_svm.mean() * 100:.2f}% ± {cv_scores_svm.std() * 100:.2f}%")
```
